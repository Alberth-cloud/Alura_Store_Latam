### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print(f"La tienda 1 tuvo un ingreso de ${ingreso_tienda:.0f}")
print(f"La tienda 2 tuvo un ingreso de ${ingreso_tienda2:.0f}")
print(f"La tienda 3 tuvo un ingreso de ${ingreso_tienda3:.0f}")
print(f"La tienda 4 tuvo un ingreso de ${ingreso_tienda4:.0f}")
```

# 2. Ventas por categoría

```python
print(tienda.groupby("Categoría del Producto")["Precio"].sum())
print(tienda2.groupby("Categoría del Producto")["Precio"].sum())
print(tienda3.groupby("Categoría del Producto")["Precio"].sum())
print(tienda4.groupby("Categoría del Producto")["Precio"].sum())
```

```python
Categorias = tienda.groupby("Categoría del Producto")["Precio"].sum()
print(Categorias)
print("Categoría más vendida de la tienda 1: ", Categorias.idxmax())
print("Categoría menos vendida de la tienda 1: ", Categorias.idxmin())

Categorias = tienda2.groupby("Categoría del Producto")["Precio"].sum()
print(Categorias)
print("Categoría más vendida de la tienda 2: ", Categorias.idxmax())
print("Categoría menos vendida de la tienda 2: ", Categorias.idxmin())


Categorias = tienda3.groupby("Categoría del Producto")["Precio"].sum()
print(Categorias)
print("Categoría más vendida de la tienda 3: ", Categorias.idxmax())
print("Categoría menos vendida de la tienda 3: ", Categorias.idxmin())


Categorias = tienda4.groupby("Categoría del Producto")["Precio"].sum()
print(Categorias)
print("Categoría más vendida de la tienda 4: ", Categorias.idxmax())
print("Categoría menos vendida de la tienda 4: ", Categorias.idxmin())

```

# 3. Calificación promedio de la tienda


```python
calificacion_tienda = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()
```

```python
print(f"La tienda 1 tuvo una calificación promedio de {calificacion_tienda:.2f}")
print(f"La tienda 2 tuvo una calificación promedio de {calificacion_tienda2:.2f}")
print(f"La tienda 3 tuvo una calificación promedio de {calificacion_tienda3:.2f}")
print(f"La tienda 4 tuvo una calificación promedio de {calificacion_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
def analizar_tienda(tienda, tienda_1):
    productos_vendidos = tienda["Producto"].value_counts()
    producto_mas_vendido = productos_vendidos.idxmax()
    producto_menos_vendido = productos_vendidos.idxmin()
    print(f"El producto más vendido es {producto_mas_vendido} y el menos vendido es {producto_menos_vendido}")
    productos_vendidos.plot(kind="bar", title=("Productos más vendidos y menos vendidos"))
    plt.show()
def analizar_tienda2(tienda2, tienda_2):
    productos_vendidos = tienda2["Producto"].value_counts()
    producto_mas_vendido = productos_vendidos.idxmax()
    producto_menos_vendido = productos_vendidos.idxmin()
    print(f"El producto más vendido es {producto_mas_vendido} y el menos vendido es {producto_menos_vendido}")
    productos_vendidos.plot(kind="bar", title=("Productos más vendidos y menos vendidos de la tienda 2"))
    plt.show()
def analizar_tienda3(tienda3, tienda_3):
    productos_vendidos = tienda3["Producto"].value_counts()
    producto_mas_vendido = productos_vendidos.idxmax()
    producto_menos_vendido = productos_vendidos.idxmin()
    print(f"El producto más vendido es {producto_mas_vendido} y el menos vendido es {producto_menos_vendido}")
    productos_vendidos.plot(kind="bar", title=("Productos más vendidos y menos vendidos de la tienda 3"))
    plt.show()
def analizar_tienda4(tienda4, tienda_4):
    productos_vendidos = tienda4["Producto"].value_counts()
    producto_mas_vendido = productos_vendidos.idxmax()
    producto_menos_vendido = productos_vendidos.idxmin()
    print(f"El producto más vendido es {producto_mas_vendido} y el menos vendido es {producto_menos_vendido}")
    productos_vendidos.plot(kind="bar", title=("Productos más vendidos y menos vendidos de la tienda 4"))
    plt.show()

```

```python
analizar_tienda(tienda, "tienda_1")
analizar_tienda2(tienda2, "tienda_2")
analizar_tienda3(tienda3, "tienda_3")
analizar_tienda4(tienda4, "tienda_4")
```
